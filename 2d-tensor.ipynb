# List to tensor

```python
a = [[11,12,13],[21,22,23]]
A= torch.tensor(a)
print(A)
print(A.ndimension()) #dimension is also called as **rank** 
print(A.shape)
print(A.size())
print(A.numel()) #no. of elements

```

**Convention is to label vertical axis 0 and horizontal axis as 1**

# Tensor <-> Numpy array

```python
to_numpy = A.numpy()
print("list: ",a)
print("2D Tensor: ",A)
print("The numpy array: ",to_numpy)
new_A = torch.from_numpy(to_numpy)

print("numpy array to tensor: ",new_A)
```

# Pandas Dataframe to tensor
  **When we pass a dictionary to pd.DataFrame(), pandas treats the keys as columns and the lists as the rows for those columns.**
  **So here a 3X2 dimensional numpy array getting created from the dictionary as there are 2 keys**

```python
df = pd.DataFrame({'a':[11,21,31], 'b':[12,22,32]})
print("Pandas Dataframe to numpy: ", df.values)
print("Type BEFORE converting: ", df.values.dtype)
new_tensor = torch.from_numpy(df.values)
print("Tensor AFTER converting: ", new_tensor)
print("Type AFTER converting: ", new_tensor.dtype)
```

```python
print(new_tensor[1,0])
print(new_tensor[1][0])
```

**Accessing element is similar as accessing it in a list e.g. getting an element at 2nd row and 3rd column is **new_tensor[1,2]****
# Indexing and slicing in a 2D Tensor

```python
print("What is the value on 1st-row first two columns? ", new_tensor[0, 0:2])
print("What is the value on 1st-row first two columns? ", new_tensor[0][0:2])
print("What is the value on 1st-column first three rows? ", new_tensor[0:3,0])
print("What is the value on 1st-column first three rows? ", new_tensor[0:3][0]) ### this is not giving the exact required one
```

But we <b>can't</b> combine using slicing on row and pick one column by using the code <code>tensor_obj[begin_row_number: end_row_number][begin_column_number: end_column number]</code>. The reason is that the slicing will be applied on the tensor first. The result type will be a two dimension again. The second bracket will no longer represent the index of the column it will be the index of the row at that time. Let us see an example. "


# Tensor Operations

```python

#matrix multiplication
#  col of A = rows in B
A = torch.tensor([[0, 1, 1], [1, 0, 1]])
B = torch.tensor([[1, 1], [1, 1], [-1, 1]])
A_times_B = torch.mm(A,B)
print("The result of A * B: ", A_times_B)
#REST operations are similar as in 1D Tensor
```
